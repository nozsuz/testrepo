## Define Graphing Function


In this section, we define the function `make_graph`. You don't have to know how the function works, you should only care about the inputs. It takes a dataframe with stock data (dataframe must contain Date and Close columns), a dataframe with revenue data (dataframe must contain Date and Revenue columns), and the name of the stock.

このセクションでは、関数make_graphを定義する。この関数がどのように動作するか知る必要はありません。この関数は、株価データのデータフレーム（データフレームには日付と終値の列が含まれていなければなりません）、収益データのデータフレーム（データフレームには日付と収益の列が含まれていなければなりません）、そして銘柄名を受け取ります。


````python
def make_graph(stock_data, revenue_data, stock):
    # Step 1: Convert 'Date' to datetime format if not already
    stock_data['Date'] = pd.to_datetime(stock_data['Date'])
    revenue_data['Date'] = pd.to_datetime(revenue_data['Date'].astype(str) + '-01-01')

    # Step 2: Apply timezone localization or removal
    # To remove timezone (making it timezone-naive)
    stock_data['Date'] = stock_data['Date'].dt.tz_localize(None)
    revenue_data['Date'] = revenue_data['Date'].dt.tz_localize(None)

    # Ensure the comparison Timestamp is also time zone-naive
    comparison_date_stock = pd.to_datetime('2021-06-14').tz_localize(None)
    comparison_date_revenue = pd.to_datetime('2021-04-30').tz_localize(None)

    
    # Filter the data up to specific dates
    stock_data_specific = stock_data[stock_data.Date <= pd.to_datetime('2021-06-14')]
    revenue_data_specific = revenue_data[revenue_data.Date <= pd.to_datetime('2021-04-30')]
    
    # Creating the subplot
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing=.3)
    
    # Adding the stock price trace
    fig.add_trace(go.Scatter(x=stock_data_specific.Date, y=stock_data_specific.Close.astype("float"), name="Share Price"), row=1, col=1)
    
    # Adding the revenue trace
    fig.add_trace(go.Scatter(x=revenue_data_specific.Date, y=revenue_data_specific.Revenue.astype("float"), name="Revenue"), row=2, col=1)
    
    # Updating axes and layout
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)
    fig.update_layout(showlegend=False, height=900, title=stock, xaxis_rangeslider_visible=True)
    
    # Displaying the figure
    fig.show()
````

## Question 2: Use Webscraping to Extract Tesla Revenue Data


Use the `requests` library to download the webpage https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm Save the text of the response as a variable named `html_data`.


````python
# URL of the webpage to download
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm'

# Send a GET request to the URL
response = requests.get(url)

# Save the text of the response
html_data = response.text
````

Parse the html data using `beautiful_soup`.


````python
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_data, 'html.parser')
````

Using `BeautifulSoup` or the `read_html` function extract the table with `Tesla Revenue` and store it into a dataframe named `tesla_revenue`. The dataframe should have columns `Date` and `Revenue`.

````python
tesla_revenue_list = pd.read_html(html_data)

tesla_revenue = tesla_revenue_list[0]

# Rename the columns to 'Date' and 'Revenue'
tesla_revenue.columns = ['Date', 'Revenue']

# Check the first few rows to confirm
print(tesla_revenue.head())
````

Execute the following line to remove the comma and dollar sign from the `Revenue` column. 

````python
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"")
print(tesla_revenue.head())
````

Execute the following lines to remove an null or empty strings in the Revenue column.


````python
# 不要な文字の除去後、空文字ではない行のみを保持
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]

# 'Revenue'列を数値型に変換。変換できない場合はNaNになる。
tesla_revenue['Revenue'] = pd.to_numeric(tesla_revenue['Revenue'], errors='coerce')

# NaN値を持つ行を削除
tesla_revenue.dropna(subset=['Revenue'], inplace=True)
````

Display the last 5 row of the `tesla_revenue` dataframe using the `tail` function. Take a screenshot of the results.


````python
print(tesla_revenue.head())
````
